Sex ratio analysis

```python
sums_per_state = census_data.groupby('State').sum()

avgs_per_state = census_data.groupby('State').mean()

```

```python
sex_ratio = avgs_per_state[['Sex.ratio..females.per.1000.males.']].round(decimals=0)
# finding the average 
mean_sex_ratio= np.full((35,1),avgs_per_state[['Sex.ratio..females.per.1000.males.']].mean()).round(0)

# states above average sex ratio
sex_ratio[sex_ratio>avgs_per_state[['Sex.ratio..females.per.1000.males.']].mean()].count()

# sex ratio for children below 6
sex_ratio_child = avgs_per_state[['Sex.ratio..0.6.years.']].round(decimals=0)


# states above average sex ratio for children below 6
sex_ratio_child[sex_ratio_child>avgs_per_state[['Sex.ratio..0.6.years.']].mean()].count()

#sex_ratio_child
```

```python


# sexratio

plt.figure(figsize=(10,20))
sns.histplot(data=sex_ratio, bins=np.arange(750,1100,25))
sns.histplot(data=sex_ratio_child, bins=np.arange(750,1100,25), palette='YlOrBr')
plt.grid()

# sns.lineplot(data=mean_sex_ratio)
plt.yticks(np.arange(0,17,1));
```

```python
fig,ax = plt.subplots(figsize=(18,6))
ratio_combined = pd.concat([sex_ratio, sex_ratio_child], axis=1)

ratio_df = ((1-(ratio_combined['Sex.ratio..0.6.years.']/ratio_combined['Sex.ratio..females.per.1000.males.']))*100).round(3)

sns.scatterplot(data=ratio_df.sort_values(ascending=True), color='blue')
plt.grid()
sns.despine()
plt.xticks(rotation=45);
```

```python
ratio_combined = pd.concat([ratio_combined,ratio_df],axis=1)
ratio_combined.rename(columns = {0:'future_sex_ratio_variation'}, inplace=True)
ratio_combined
```

```python
# finding out how many states have made positive progress in their sex ratio in generations
ratio_combined[ratio_combined.future_sex_ratio_variation>0].future_sex_ratio_variation.count()
```

Eduation Analysis

```python
# Literacy
literacy_df = census_data[['State','Persons..literate',
       'Males..Literate', 'Females..Literate', 'Persons..literacy.rate',
       'Males..Literatacy.Rate', 'Females..Literacy.Rate']]
literacy_sums = literacy_df.iloc[:,[0,1,2,3]].groupby('State').sum()
literacy_avgs = literacy_df.iloc[:,[0,4,5,6]].groupby('State').mean()

```

```python
literacy_sums.sort_values(by='Persons..literate')
```

```python
# Literate people graph
fig, axs = plt.subplots(figsize=(18,18))
temp = np.arange(len(literacy_sums['Males..Literate']))

plt.bar(temp -0.2, literacy_sums['Females..Literate'], width=0.4, label = 'Female')
plt.bar(temp +0.2, literacy_sums['Males..Literate'], width=0.4, label = 'Male')
plt.ylabel('Literate population')
plt.xlabel('States')
plt.title('Literate population by gender')
plt.xticks(temp,literacy_sums.index.values, rotation=55)

;
```

In all of the states the number of male literates is greater than that of females.

Kerela is the only state to have an almost equal number of Male to Female literates. ie 12753602 males to 12732086 females.

```python
percent = (literacy_sums['Males..Literate'].sum()/(literacy_sums['Persons..literate'].sum()))*100

print("As of 2001 there are {} literate males vs {} literate females. Therefore there is a disparity of {} more males who are literate as compared to women".format(literacy_sums['Males..Literate'].sum(),literacy_sums['Females..Literate'].sum(),(literacy_sums['Males..Literate'].sum() -literacy_sums['Females..Literate'].sum())))
print(f"{percent.round(2)}% of the total literate population is male while the rest is female.")
# .format(literacy_sums['Males..Literate'].sum(),literacy_sums['Females..Literate'].sum(),(literacy_sums['Males..Literate'].sum() -literacy_sums['Females..Literate'].sum()))


```

UP has the largest number of literate people (75719284), 48901413 of which are male and 26817871 female

Lakshdweep has the lowest number of literate people (44683), 24511 of which are male and 20172 female.

Analysis of education category

```python
education_sums = sums_per_state[['Total.Educated',
       'Data.without.level', 'Below.Primary', 'Primary', 'Middle',
       'Matric.Higher.Secondary.Diploma', 'Graduate.and.Above']]

```

```python
ax, fig = plt.subplots(figsize=(18,6))

sns.barplot(data=education_sums[['Data.without.level', 'Below.Primary', 'Primary', 'Middle',
       'Matric.Higher.Secondary.Diploma', 'Graduate.and.Above']])
```

```python
total = int(education_sums['Total.Educated'].sum())
below_primary = ((education_sums['Below.Primary'].sum()/total)*100).round(1)
primary = ((education_sums['Primary'].sum()/total)*100).round(1)
middle = ((education_sums['Middle'].sum()/total)*100).round(1)
matric_Hsc_diploma = ((education_sums['Matric.Higher.Secondary.Diploma'].sum()/total)*100).round(1)
graduate_above = ((education_sums['Graduate.and.Above'].sum()/total)*100).round(1)
no_data = ((education_sums['Data.without.level'].sum()/total)*100).round(1)
```

```python
print(f'Out of the total of {total} number of people surveyed , {below_primary}% have below primary level education. {primary}% have primary level, {middle}% have middle level, {matric_Hsc_diploma}% have matric/Higher secondary/ diploma, while only {graduate_above}% of the people are Graduates or above. {no_data}% of the people either had no level of education')

```

About 50% of the population has either a primary or secondary level of education.

About 70% of the population has accesessed at least the basic level of education i.e below matric. 
While the rest 30% have been privy to higher educaiton.

```python
education_sums['certificated'] = (((education_sums['Matric.Higher.Secondary.Diploma']+education_sums['Graduate.and.Above'])/education_sums['Total.Educated'])*100).round(2)

ax, fig = plt.subplots(figsize=(18,6))

sns.barplot(data=education_sums, x=education_sums.index.values, y=education_sums.certificated )
plt.xticks(rotation=55);

education_sums.certificated.sort_values()
                                                                                                             
```

Working population analysis

```python
working_sums = sums_per_state[['X0...4.years',
       'X5...14.years', 'X15...59.years', 'X60.years.and.above..Incl..A.N.S..',
       'Total.workers', 'Main.workers', 'Marginal.workers', 'Non.workers']]
```

Here just using the categories already provided we will assume that the columns of X15..59 are the working population and the others are dependant population

```python
fig, ax = plt.subplots(1,2,figsize=(18,6))

# age group analysis 
labels = ['0-4 years','5-14 years','15-59 years', '60 years and above' ]
explode = [.1,.12,.05,.23]

ax[0].pie(working_sums.iloc[:,:4].sum(axis=0), 
        labels=labels, 
        explode=explode, 
        autopct='%1.1f%%',
       shadow=True)

# categorisation of working population
labels1 = ['Main workers','Marginal workers','Non workers']
ax[1].pie(working_sums.iloc[:,5:].sum(axis=0), 
        labels=labels1, 
        explode=explode[:3], 
        autopct='%1.1f%%',
       shadow=True)
;
```

```python
fig, ax = plt.subplots(2,1,figsize=(18,25))

working_sums['dependant_population'] = (working_sums['X5...14.years'] + working_sums['X0...4.years']+working_sums['X60.years.and.above..Incl..A.N.S..'])
working_sums.rename(columns={'X15...59.years':'working_potential_population'}, inplace=True)

working_sums.plot(y=['working_potential_population','dependant_population'], kind="bar", ax=ax[0], rot=55)

working_categories=working_sums.iloc[:,5:-1]
working_categories
working_categories.columns = ['Main workers','Marginal workers','Non workers']
working_categories.plot(y=working_categories.columns, kind="bar", ax=ax[1],rot=55);

```

```python
(working_sums['working_potential_population']/working_sums['dependant_population']).round(2).sort_values(ascending=False)
working_people = working_sums['working_potential_population'].sum()
dependant_people = working_sums['dependant_population'].sum()
```

This shows that most of the population falls in the 15 to 59 years range which is the population with working age (56.9%).

UP has the highest working population (85888210) and Lakshwadeep has the lowest (36134)

The union territory of Daman and Diu has the highest rate of potentially working people i.e people in the age group of 15 to 60.
they have more than 2 potentially working persons for every dependant person.

Bihar has the least number of potentially working people, with almost equal number of protentially working and dependant persons.

It is a good sign that there are more potentially working people as compared to the dependant population.

```python
print(f'There are totally {working_people} petentially working people compared to {dependant_people} dependant persons. \n This is means there is on average {((working_people/dependant_people)*100).round(0)} working people for every 100 dependant people in India.')

```

About 61% of the working population is composed of non workers, which is a bit worrisome.
Also only about 30% of the potentially working people are main workers while the rest are marginal workers.

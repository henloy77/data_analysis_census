Sex ratio analysis

```python
sums_per_state = census_data.groupby('State').sum()

avgs_per_state = census_data.groupby('State').mean()

```

```python
sex_ratio = avgs_per_state[['Sex.ratio..females.per.1000.males.']].round(decimals=0)
# finding the average 
mean_sex_ratio= np.full((35,1),avgs_per_state[['Sex.ratio..females.per.1000.males.']].mean()).round(0)

# states above average sex ratio
sex_ratio[sex_ratio>avgs_per_state[['Sex.ratio..females.per.1000.males.']].mean()].count()

# sex ratio for children below 6
sex_ratio_child = avgs_per_state[['Sex.ratio..0.6.years.']].round(decimals=0)


# states above average sex ratio for children below 6
sex_ratio_child[sex_ratio_child>avgs_per_state[['Sex.ratio..0.6.years.']].mean()].count()

#sex_ratio_child
```

```python


# sexratio

plt.figure(figsize=(10,20))
sns.histplot(data=sex_ratio, bins=np.arange(750,1100,25))
sns.histplot(data=sex_ratio_child, bins=np.arange(750,1100,25), palette='YlOrBr')
plt.grid()

# sns.lineplot(data=mean_sex_ratio)
plt.yticks(np.arange(0,17,1));
```

```python
fig,ax = plt.subplots(figsize=(18,6))
ratio_combined = pd.concat([sex_ratio, sex_ratio_child], axis=1)

ratio_df = ((1-(ratio_combined['Sex.ratio..0.6.years.']/ratio_combined['Sex.ratio..females.per.1000.males.']))*100).round(3)

sns.scatterplot(data=ratio_df.sort_values(ascending=True), color='blue')
plt.grid()
sns.despine()
plt.xticks(rotation=45);
```

```python
ratio_combined = pd.concat([ratio_combined,ratio_df],axis=1)
ratio_combined.rename(columns = {0:'future_sex_ratio_variation'}, inplace=True)
ratio_combined
```

```python
# finding out how many states have made positive progress in their sex ratio in generations
ratio_combined[ratio_combined.future_sex_ratio_variation>0].future_sex_ratio_variation.count()
```

```python
avgs_per_state[['Sex.ratio..0.6.years.']].mean()
```

Only 3 states have a positive sex ratio i.e more females than males (Uttranchal, Pondicherry, Kerala)

```python
# urban and rural
rural_urban_pop = census_data.groupby('State').count()[['Rural', 'Urban']]
rural_urban_pop['difference'] = rural_urban_pop.Rural -rural_urban_pop.Urban

rural_percent = (rural_urban_pop.Rural/(rural_urban_pop.Rural+rural_urban_pop.Urban))*100
rural_urban_pop.Rural.sum()

f,ax = plt.subplots(figsize=(18,6))
plt.xticks(rotation=65);
sns.barplot(y=rural_urban_pop.difference.sort_values(ascending=True), x = rural_urban_pop.index, ax= ax);
```

```python
f,ax = plt.subplots(figsize=(18,6))

sns.lineplot(y=rural_urban_pop.Rural.sort_values(ascending=True), x = rural_urban_pop.index)
sns.lineplot(y=rural_urban_pop.Urban.sort_values(ascending=True), x = rural_urban_pop.index)
plt.xticks(rotation=65);
```

```python
# Literacy
literacy_df = census_data[['State','Persons..literate',
       'Males..Literate', 'Females..Literate', 'Persons..literacy.rate',
       'Males..Literatacy.Rate', 'Females..Literacy.Rate']]
literacy_sums = literacy_df.iloc[:,[0,1,2,3]].groupby('State').sum()
literacy_avgs = literacy_df.iloc[:,[0,4,5,6]].groupby('State').mean()

```

```python
# Literate people graph
fig, axs = plt.subplots(2, 2,figsize=(18,18))

sns.barplot(data=literacy_sums, ax=axs[0])
axs[0].set_ylabel('Count')
axs[0].set_xlabels(['Persons', 'Males', 'Females'])
;
```

```python
# Literate rate graph
plt.figure(figsize=(18,6))
sns.barplot(data=literacy_avgs)

plt.xlabel('Average Literacy')
plt.ylabel('Percent')
;
```
